A análise das correlações vai trazer algumas comparações entre os parametros. A correlação é uma medida que vai de -1 0 +1.

Como interpretar:
1. Quando um valor é igual a zero é porque um parametro não tem nenhuma relação com o outro, também chamado de relação FRACA.
2. Quando o valor é igual mais próximo dos extremos (-1 e +1) existe uma correlação FORTE entre as variavéis.
3. Quando aparece um valor NEGATIVO é porque elas são inversamente proporcionais, ou seja, quando uma sobe a outra desce, na proporção do resultado.
4. Quando aparece um valor positivo é porque elas são deiratemente proporcionais quando uma sobe a outra também sobe, na proporção do resultado.

Analisando os resultados é possível ver, por exemploe que: 
> A relação mais FORTE entre NOTAS é NU_NOTA_CH e NU_NOTA_LC, com resultado de  0.744170

###METODO CROSS VALIDATION

O metodo serve para comparar modelos. Por isso ele retorna, por padrão o valor negativo (Principio do 'Quanto masi alto melhor'), e retorna o tempo para poder comparar o desempenho dos modelos.

```
from sklearn.model_selection import cross_validate 

modelo_arvore = DecisionTreeRegressor(max_depth=2)
cross_validate(modelo_arvore, x, y)

```

O Crooss validate retornou as informaçõeos 'fit-time, score_time e test_score. 

Os valores da variação ('0.4407387') é bem inferior ao modelo que apresentou valor de 6075.986466..... Por que?:
Porque o cross_validate possui um parametro cahamado "scoring" , que insere o metodo de avaliação que queremos trabalhar.  

Obs.  Não informamos quantos splits queriamos. Por padrão o metodo utilizou 5 splits dos dados..


```
from sklearn.model_selection import cross_validate 

modelo_arvore = DecisionTreeRegressor(max_depth=2)
cross_validate(modelo_arvore, x, y, scoring='neg_mean_squared_error')
```

Agora a variação mostrada é bem parecida com o modelo anterior, mas está negativo...
Isso acontece por causa da metrica interna do algoritmo.. ('mais alto melhor').

Vamos incluir no calculo um limite maxio e minio (Intervalod e confiança)

```
from sklearn.model_selection import cross_validate 

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv=10, scoring='neg_mean_squared_error')
media = (resultados['test_score']*-1).mean()
desvio_padrao = (resultados['test_score']*-1).std()
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f'Intervalo de confiança {lim_inferior} - {lim_superior}')
```

Criando uma fuinção para receber o calculo da média com limites:


```
def calcula_mse(resultados):
    media = (resultados['test_score']*-1).mean()
    desvio_padrao = (resultados['test_score']*-1).std()
    lim_inferior = media - (2*desvio_padrao)
    lim_superior = media + (2*desvio_padrao)
    print(f'Intervalo de confiança {lim_inferior} - {lim_superior}')
```
